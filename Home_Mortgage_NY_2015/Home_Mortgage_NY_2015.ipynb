```python
features= [i for i in ny_hmda_2015_dataset.columns if ny_hmda_2015_dataset[i].dtype != 'object']
```

```python
list_to_remove = ['action_taken','purchaser_type', 'denial_reason_1','denial_reason_2',
                  'denial_reason_3','sequence_number']

features= list(set(features).difference(set(list_to_remove)))

X = ny_hmda_2015_dataset[features]
y = ny_hmda_2015_dataset['action_taken']
```

```python
def change_action_taken(y):
    if ( y == 1):
        return 1
    else:
        return 0
```

```python
y = ny_hmda_2015_dataset['action_taken'].apply(change_action_taken)
X = X.fillna(0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model_logistic= LogisticRegression()
model_logistic.fit(X_train,y_train)
```

```python
pred_logistic=model_logistic.predict(X_test)
logistic_score =accuracy_score(pred_logistic, y_test)
logistic_pred_score =round(logistic_score*100,2)
logistic_pred_score
```

```python
print(classification_report(y_test,pred_logistic))
```

### CatBoostClassifier

```python
from catboost import CatBoostClassifier

model_catboost=CatBoostClassifier(loss_function='MultiClassOneVsAll',random_seed = 0)
model_catboost.fit(X_train,y_train, verbose = 1000)
```

```python
pred_catboost=model_catboost.predict(X_test)
catboost_score =accuracy_score(pred_catboost, y_test)
catboost_pred_score =round(catboost_score*100,2)
catboost_pred_score
```

```python
print(classification_report(y_test,pred_catboost))
```

### RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

model_randomforest= RandomForestClassifier()
model_randomforest.fit(X_train,y_train)
```

```python
pred_randomforest=model_randomforest.predict(X_test)
randomforest_score =accuracy_score(pred_randomforest, y_test)
randomforest_pred_score =round(randomforest_score*100,2)
randomforest_pred_score
```

```python
print(classification_report(y_test,pred_randomforest))
```

### xgboost

```python
import xgboost as xgb

model_xgboost= xgb.XGBClassifier()
model_xgboost.fit(X_train,y_train)
```

```python
pred_xgboost=model_xgboost.predict(X_test)
xgboost_score =accuracy_score(pred_xgboost, y_test)
xgboost_pred_score =round(xgboost_score*100,2)
xgboost_pred_score
```

```python
print(classification_report(y_test,pred_xgboost))
```

### LGBMClassifier

```python
import lightgbm as lgb

model_lightgbm = lgb.LGBMClassifier()
model_lightgbm.fit(X_train,y_train)
```

```python
pred_lightgbm=model_lightgbm.predict(X_test)
lightgbm_score =accuracy_score(pred_lightgbm, y_test)
lightgbm_pred_score =round(lightgbm_score*100,2)
lightgbm_pred_score
```

```python
print(classification_report(y_test,pred_lightgbm))
```

```python
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
def print_evaluate(true, predicted):  
    mae = metrics.mean_absolute_error(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    r2_square = metrics.r2_score(true, predicted)
    print('MAE:', mae)
    print('MSE:', mse)
    print('RMSE:', rmse)
    print('R2 Square', r2_square)
```

```python
print_evaluate(y_train,model_xgboost.predict(X_train))
```
