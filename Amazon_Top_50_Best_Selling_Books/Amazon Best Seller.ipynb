```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
original_amazon_bestsellers_dataset = pd.read_csv('bestsellers with categories.csv')
amazon_bestsellers_dataset = original_amazon_bestsellers_dataset.copy()
amazon_bestsellers_dataset.head()
```

```python
amazon_bestsellers_dataset.info()
```

```python
amazon_bestsellers_dataset.isnull().sum()
```

```python
amazon_bestsellers_dataset.Genre.unique()
```

```python
amazon_book_genre_data = [len(amazon_bestsellers_dataset[amazon_bestsellers_dataset['Genre']=='Fiction']),
                    len(amazon_bestsellers_dataset[amazon_bestsellers_dataset['Genre']=='Non Fiction'])]
amazon_book_genre_labels = ['Fiction', 'Non Fiction']
plt.title('Book Genre')
plt.pie(amazon_book_genre_data, labels = amazon_book_genre_labels, autopct='%.0f%%')
plt.show()
```

```python
sns.displot(amazon_bestsellers_dataset['User Rating'])
```

```python
sorted_by_User_rating_dataset = amazon_bestsellers_dataset.sort_values(['User Rating'],ascending=False)
sorted_by_User_rating_dataset
```

### List of Amazon best seller with User Rating 4.9 over the year (2009 - 2019)
In 2009 and 2010, it doesn't have any book that have rating 4.9.

```python
sorted_by_User_rating_dataset[sorted_by_User_rating_dataset['User Rating']==4.9].sort_values(['Year'])
```

```python
fig, axs = plt.subplots(nrows=3, figsize=(8,8))
g1 = sns.regplot(x="Price", 
                 y="User Rating", 
                 data =sorted_by_User_rating_dataset,
                 ax=axs[0]).set_title("User Rating vs. Price")
g2 = sns.regplot(x="Price", 
                 y="Reviews", 
                 data =sorted_by_User_rating_dataset,
                 ax=axs[1]).set_title("User Rating vs. Price")
g3 = sns.regplot(x="Reviews", 
                 y="User Rating",
                 data =sorted_by_User_rating_dataset,
                 ax=axs[2]
                ).set_title("User Rating vs. Reviews")
fig.tight_layout(pad=3.0)
```

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.boxplot(x="Year", y="User Rating",
            hue="Genre", palette=["m", "g"],
            data=amazon_bestsellers_dataset).set_title("Amazon Best Selling Books 2009-2019 by Genre")
```
