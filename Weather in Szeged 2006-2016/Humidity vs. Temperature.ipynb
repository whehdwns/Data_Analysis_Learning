Is there a relationship between humidity and temperature?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
weather_data=pd.read_csv('weatherHistory.csv')
weather_data
```

```python
weather_data.describe()
```

```python
weather_data.columns
```

```python
humanity_temp=weather_data[['Temperature (C)','Humidity']]
humanity_temp
```

```python
humanity_temp.describe()
```

```python
humanity_temp.isnull().sum()
```

```python
sns.pairplot(humanity_temp)
```

```python
sns.jointplot(data=humanity_temp, x='Humidity', y='Temperature (C)')
```

```python
X=humanity_temp[['Humidity']]
Y=humanity_temp[['Temperature (C)']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(normalize=True)
model.fit(X_train,Y_train)
```

```python
model.intercept_
```

```python
pred = model.predict(X_test)
print(pred)
```

```python
model.score(X_test,Y_test)
```

```python
plt.scatter(X_test, Y_test)
plt.plot(X_test, pred, color = 'g')
plt.title("Relationship between humidity and temperature")
plt.xlabel("Humidity")
plt.ylabel("Temperature (C)")
```

```python
sns.distplot((Y_test-pred),bins=100);
```

```python
test_pred = model.predict(X_test)
train_pred = model.predict(X_train)
```

```python
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
def print_evaluate(true, predicted):  
    mae = metrics.mean_absolute_error(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    r2_square = metrics.r2_score(true, predicted)
    print('MAE:', mae)
    print('MSE:', mse)
    print('RMSE:', rmse)
    print('R2 Square', r2_square)
```

```python
print_evaluate(Y_train,train_pred)
```
